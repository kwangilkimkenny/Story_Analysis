```python
import spacy

nlp = spacy.load("en_core_web_lg")
doc = nlp(academicVerbs)
```

```python
print(f"{'text':{8}} {'POS':{6}} {'TAG':{6}} {'Dep':{6}} {'POS explained':{20}} {'tag explained'} ")

for token in doc:
    print(f'{token.text:{2}} {","} {token.pos_:{6}} {","} {token.tag_:{6}}  {","} {token.dep_:{6}}  {","} {spacy.explain(token.pos_):{20}} {","} {spacy.explain(token.tag_)}')
```

```python
print(f"{'text':{8}} {", ":{1}} {'POS':{6}}  {", ":{1}} {'TAG':{6}}  {", ":{1}} {'Dep':{6}} {", ":{1}}  {'POS explained':{20}}  {", ":{1}} {'tag explained'} ")

for token in doc:
    print(f'{token.text:{8}} {","} {token.pos_:{6}} {","} {token.tag_:{6}}  {","} {token.dep_:{6}}  {","} {spacy.explain(token.pos_):{20}} {","} {spacy.explain(token.tag_)}')
```

```python
p_tag_list = []
for token in doc:
    p_tag_list.append((token.text, token.pos_, token.tag_, token.dep_, spacy.explain(token.pos_),spacy.explain(token.tag_)))
    
```

```python
import pandas as pd
data = pd.DataFrame(p_tag_list, columns = ['word', 'POS', 'TAG', "Dep", "Pos explained","Tag explained"])
data
```
