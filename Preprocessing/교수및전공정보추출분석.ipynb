```python
import os
import numpy as np
import pandas as pd
from IPython import display

pd.options.display.max_rows = 999 # 데이터 프레임 표시 최대 열수를 999로 지정
pd.set_option('display.max_columns',999) # 데이터 프레임 표시 최대 행수를 999로 지정
```

```python
# get excel data to dataframe
get_file = './college_datasets/UC Berkeley/Professors 전공정보 및 General 대학 정보 링크 - UC Berkeley.csv'

df = pd.read_csv(get_file)
df
```

```python
# 소문자로 변환
df['Professor'] = df['Professor'].astype(str).str.lower()
df['First Major'] = df['First Major'].astype(str).str.lower()
df['Second Major'] = df['Second Major'].astype(str).str.lower()
df['Third Major'] = df['Third Major'].astype(str).str.lower()
df['Fourth Major'] = df['Fourth Major'].astype(str).str.lower()
df['Fifth Major'] = df['Fifth Major'].astype(str).str.lower()
df['General Info/Bio, Research & Publications'] = df['General Info/Bio, Research & Publications'].astype(str).str.lower()
df['Other Academic information'] = df['Other Academic information'].astype(str).str.lower()
df
```
