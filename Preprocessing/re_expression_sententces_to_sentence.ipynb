```python
# 문장들을 문장리스트로 분해하고 다시 각 문장을 단어 리스트로 분해하여 이중리스트를 만드는 Method
def sents_to_sent_list(sentence_lower):
    wdt_sents = []
    for i in sentence_lower:
        wdt = word_tokenize(i)
        wdt_sent = []
        for j in wdt:
            lw_wdt = j.lower() # 소문자 변환
            wdt_sent.append(lw_wdt)
        wdt_sents.append(wdt_sent)

    return wdt_sents
```
