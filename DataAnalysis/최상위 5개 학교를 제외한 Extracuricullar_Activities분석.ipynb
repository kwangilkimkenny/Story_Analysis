```python
def get_hotwords(text):
    result = []
    pos_tag = ['PROPN', 'ADJ', 'NOUN'] # 1
    doc = nlp(text.lower()) # 2
    for token in doc:
        # 3
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
        # 4
        if(token.pos_ in pos_tag):
            result.append(token.text)
                
    return result # 5
```

```python
top_college = ['[harvard-university-cambridge-ma]', 
    '[yale-university-new-haven-ct]', 
    '[stanford-university-stanford-ca]', 
    '[massachusetts-institute-of-technology-cambridge-ma]',
    '[university-of-pennsylvania-philadelphia-pa]',
    '[california-institute-of-technology-pasadena-ca]',
    '[brown-university-providence-ri]']
```

```python
## 정규식 이용 하여 []안의 단어를 추출한다. 나중에 숫자를 제거한다.
def get_ext_data(input_text):
    p = re.compile(r"\[(.+)\](.+)")
    try:
        m = p.match(input_text + " ")
        ext_result = m.group(1)
    except:
        ext_result = ""
    
    return ext_result
```

```python
from tqdm import tqdm

with open("./DataEssay/essayData.json") as f:
    data = json.load(f)

get_all_result__ = []

for i in tqdm(range(len(data["data"]))):
    
    data_str = data["data"][i]["undergrad_schools_applied"]
    words_list = data_str.split()
    if words_list[1] in top_college:
        pass
    else:
        # top_college 에 Erolled 학교가 있다면, 해당 데이터의  Extracurricular_activite를 추출한다. 이부분 함수로 처리할 것
        data_ext_curr = data["data"][i]["extracurricular_activitie"]
        print("essay no:", i)
        print("------------------------------------------------------")
        print(data_ext_curr)
        print("------------------------------------------------------")
        words_list_ = data_ext_curr.split()
        print("words_list_ :", words_list_)
        print("------------------------------------------------------")
        for j in words_list_:
            print('j :', j)
            result = get_ext_data(j)
            print("result :", result)
            #숫자인지 판단하고, 숫자가 아니면 append
            if result.isdigit():
                pass
            else:
                get_all_result__.append(result)
```
